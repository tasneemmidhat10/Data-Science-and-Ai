```python
X = data_1.drop(['Expert Diagnose'], axis=1)
Y = data_1['Expert Diagnose']
X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.4, random_state = 0)
# When I set the value of the random_state to 0, the model's accuracy is 0.8958. Whem random_state = 42 the model's accuracy is 0.75
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
y_predict = clf.predict(x_test)
print('Accurcy :', accuracy_score(y_test, y_predict))
```

```python
para_grid = {'max_depth' : [10,20,30, None], 'min_samples_split' : [2,5,10], 'min_samples_leaf' : [1, 2, 4], 'max_features' : ['sqrt', 'log2', None]}
grid_search = GridSearchCV(estimator = clf, param_grid = para_grid, cv = 5, n_jobs = -1, verbose = 2, scoring = 'accuracy')
grid_search.fit(x_train, y_train)
clf_best = grid_search.best_estimator_
```

```python
def pipeline(columns):
    import pandas as pd
    patient_name = input(' Hello please Enter your patient`s name :')
    patient_record = [[]]
    dic = {'seldom' : 0, 'sometimes' : 1, 'usually' : 2, 'most-often' : 3}
    sadness = input(' How often do you feel sad (Most_times-Usually-Sometimes-Seldom)')
    patient_record.append(sadness)
    euphoric = input(' How often do you feel euphoria ? (Most-times_Usually- Sometimes- Seldom)')
    patient_record.append(euphoric)
    exhuasted = input('How often do you feel exhasuted ? (Most-times-Uusally-Sometimes-Seldom)')
    patient_record.append(exhuasted)
    Sleep_disorder = input(' How often is your sleep pattern disrupted? (Most-times, Usually, Sometimes-Seldom)')
    patient_record.append(Sleep_disorder)
    mood_swing = input('Do you have a mood swings? (YES, NO)')
    patient_record.append(mood_swing)
    suicidal_thoughts = input('Do you have suicidal thoughts? (YES, NO)')
    patient_record.append(suicidal_thoughts)
    anorexia = input('Do you have Anorexia?(YES, No)')
    patient_record.append(anorexia)
    authority_respect = input(' Do you respect authority and authority figures? (YES, NO)')
    patient_record.append(authority_respect)
    try_expl = input(' Do you do Try & explanation ? (YES, NO)')
    patient_record.append(try_expl)
    aggr_res = input(' Are your responses aggressive ? (YES, NO)')
    patient_record.append(aggr_res)
    ignore_move_on = input(' Do you ignore what has happened to you and move_on ?(YES,NO)')
    patient_record.append(ignore_move_on)
    ner_breakdown = input('Do you often get nervous break downs? (YES , NO)')
    patient_record.append(ner_breakdown)
    admit_mis = input('Do you admit your mistakes?(YES, NO)')
    patient_record.append(admit_mis)
    ovr_th = input('Do you overthink ? (YES, NO)')
    patient_record.append(ovr_th)
    sexual_act = input(' On a scale from 0-10 describe your sexual activity? ')
    patient_record.append(sexual_act)
    concent = input(' On a scale from 0-10 describe your concentration?')
    patient_record.append(concent)
    optim = input('On a scale from 0-10 describe your optimism level : ')
    patient_record.append(optim)
    df = pd.DataFrame(patient_record, columns = columns)
    return df
```
